```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import pairwise_distances_argmin_min
```

```python
data = pd.read_csv("Datasets/q1.csv")
```

```python
X = data[['x', 'y']].values
```

```python
def plot_clusters(X, centers, labels, title):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.5)
    plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, marker='X')
    plt.title(title)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.show()
```

```python
def k_means(X, n_clusters, max_iter=100):
    # Randomly initialize centroids
    centers = X[np.random.choice(len(X), n_clusters, replace=False)]
    for _ in range(max_iter):
        # Assigning each data point to the nearest centroid
        labels = pairwise_distances_argmin_min(X, centers)[0]
        # Updating centroids
        new_centers = np.array([X[labels == i].mean(axis=0) for i in range(n_clusters)])
        # If no points have changed clusters, convergence is reached
        if np.all(centers == new_centers):
            break
        centers = new_centers
    return centers, labels
```

```python
from sklearn.metrics import pairwise_distances


def k_medoids(X, n_clusters, max_iter=100):
    if n_clusters > len(X):
        raise ValueError("Number of clusters cannot be greater than the number of data points.")

    # Randomly initialize medoids
    medoids_idx = np.random.choice(len(X), n_clusters, replace=False)
    medoids = X[medoids_idx]
    
    for _ in range(max_iter):
        # Assign each data point to the nearest medoid
        distances = pairwise_distances(X, medoids)
        labels = np.argmin(distances, axis=1)
        
        # Update medoids
        new_medoids = np.array([np.mean(X[labels == i], axis=0) for i in range(n_clusters)])
        new_distances = pairwise_distances(X, new_medoids)
        new_labels = np.argmin(new_distances, axis=1)
        
        # If no medoids have changed, convergence is reached
        if np.array_equal(labels, new_labels):
            break
        
        medoids = new_medoids
        labels = new_labels
    
    return medoids, labels
```

```python

# Plot initial data
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], s=50, alpha=0.5)
plt.title('Initial Data')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Perform K-means clustering
n_clusters=2
centers, labels = k_means(X, n_clusters)
class_counts = {f"Class {i}": np.sum(labels == i) for i in range(n_clusters)}
print("Class Counts after K-means clustering:", class_counts)
plot_clusters(X, centers, labels, 'K-means Clustering (Initial)')
```

```python
# Perform K-medoids clustering
medoids, labels = k_medoids(X, n_clusters)
class_counts = {f"Class {i}": np.sum(labels == i) for i in range(n_clusters)}
print("Class Counts after K-medoids clustering:", class_counts)
plot_clusters(X, medoids, labels, 'K-medoids Clustering (Initial)')
```
